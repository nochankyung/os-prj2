```python
import pandas as pd
```

```python
df = pd.read_csv('2019_kbo_for_kaggle_v2.csv')
```

```python
for year in range(2015, 2019):
    year_data = df[df['year'] == year].copy()
 
    # hits(안타), batting average(타율), homerun(홈런), onbase percentage(출루율)에 대해 순위 매기기
    year_data['hits_rank'] = year_data['H'].rank(method='max', ascending=False)
    year_data['avg_rank'] = year_data['avg'].rank(method='max', ascending=False)
    year_data['hr_rank'] = year_data['HR'].rank(method='max', ascending=False)
    year_data['obp_rank'] = year_data['OBP'].rank(method='max', ascending=False)

    # 각 통계에서 상위 10명 선택
    top_10_hits = year_data[year_data['hits_rank'] <= 10].sort_values('hits_rank')
    top_10_avg = year_data[year_data['avg_rank'] <= 10].sort_values('avg_rank')
    top_10_hr = year_data[year_data['hr_rank'] <= 10].sort_values('hr_rank')
    top_10_obp = year_data[year_data['obp_rank'] <= 10].sort_values('obp_rank')

    # hits(안타), batting average(타율), homerun(홈런), onbase percentage(출루율)에서 상위 10명 출력
    top_10_hits = year_data[year_data['hits_rank'] <= 10].sort_values('hits_rank')
    top_10_avg = year_data[year_data['avg_rank'] <= 10].sort_values('avg_rank')
    top_10_hr = year_data[year_data['hr_rank'] <= 10].sort_values('hr_rank')
    top_10_obp = year_data[year_data['obp_rank'] <= 10].sort_values('obp_rank')
    
    print(f"\nTop 10 in Hits({year}):")
    print(top_10_hits[['batter_name', 'H']])

    print(f"\nTop 10 in Batting Average({year}):")
    print(top_10_avg[['batter_name', 'avg']])

    print(f"\nTop 10 in Homerun({year}):")
    print(top_10_hr[['batter_name', 'HR']])

    print(f"\nTop 10 in On-base Percentage({year}):")
    print(top_10_obp[['batter_name', 'OBP']])
```

```python
data_2018 = df[df['year'] == 2018]

# 각 포지션별로 승리 기여도가 가장 높은 선수 찾기
position_rank = data_2018.groupby('cp')['war'].rank(method='max', ascending=False)

# 가장 높은 승리 기여도를 가진 선수 출력
max_war_positions = data_2018[position_rank == 1]
print("Player with the highest WAR by position in 2018:")
print(max_war_positions[['cp', 'batter_name', 'war']])
```

```python
# 필요한 열 선택하여 상관 관계 계산
selected_columns = ['R', 'H', 'HR', 'RBI', 'SB', 'war', 'avg', 'OBP', 'SLG', 'salary']
correlation = df[selected_columns].corr()['salary'].sort_values(ascending=False)

# 상관 관계 출력
print("Correlation between selected features and salary:")
print(correlation.iloc[1:].to_string())
print("\nHighest correlation with salary:")
print(correlation.iloc[1:].head(1).to_string())
```
